## Outlier detection

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
df[['Occupation','Product_Category_1','Product_Category_2','Age','Product_Category_2','Product_Category_3']].plot(kind='box')
plt.xticks(rotation=90)
```

```python
def clif(x):
    ul = x.quantile(0.75)+1.5*(x.quantile(0.75)-x.quantile(0.25))
    ll = x.quantile(0.25)-1.5*(x.quantile(0.75)-x.quantile(0.25))
    x = x.clip_upper(ul)
    x = x.clip_lower(ll)
    return x

df[['Product_Category_1']]=df[['Product_Category_1']].apply(lambda x:clif(x))
```

```python
df[['Occupation','Product_Category_1','Product_Category_2','Age','Product_Category_2','Product_Category_3']].plot(kind='box')

plt.rcParams['figure.figsize'] = 5,5
plt.xticks(rotation=90)
```
